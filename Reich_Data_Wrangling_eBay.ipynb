### Ram Size Column

```python
# get list of unique values for Ram column
ram_unique_values = pc_df['Ram Size'].unique()
print(f"There are {len(ram_unique_values)} unique Ram Size values: {ram_unique_values}")
```

```python
# Standardize RAM Size with help from ChatGPT
# Citation: ChatGPT. (2025, September 9). Conversation with ChatGPT about cleaning and standardizing RAM size values in Python/pandas. OpenAI. Personal communication.

import re

def map_ram(ram):
    # Address missing and unknown values
    if pd.isna(ram):
        return "Unknown"
    r = str(ram).lower().strip()

    if any(x in r for x in ["unknown", "not included"]):
        return 'Unknown'
    
    # If RAM value has a number, create a RAM size local variable and extract the digits only;
    #    otherwise, label it as 'unknown' because we cannot make conjecture about an unknown value here.
    match = re.search(r"(\d+)", r)
    if not match:
        return "Unknown"
    size = int(match.group(1))

    # Standardize to only GB values: convert from MB to GB: 1024 MB = 1 GB
    if "mb" in r:
        size = round(size / 1024)
        # Handle zero values: We cannot know how much RAM if the value is 0, so we label as 'Unknown'
        if size == 0:
            return "Unknown"

    # Standardize the RAM values and return as a properly formatted GB string value
    if "64" in r:
        size = 64
    elif "40" in r:
        size = 40
    elif "36" in r:
        size = 36
    elif "32" in r:
        size = 32
    elif "24" in r:
        size = 24
    elif "20" in r:
        size = 20
    elif "16" in r:
        size = 16
    elif "12" in r:
        size = 12
    elif "8" in r:
        size = 8
    elif "6" in r:
        size = 6
    elif "4" in r:
        size = 4
    elif "3" in r:
        size = 3
    elif "2" in r:
        size = 2
    elif "1" in r:
        size = 1
    elif "128" in r:
        size = 128
    elif "256" in r:
        size = 256
    elif "512" in r:
        size = 512

    return f"{size} GB"

# Map the result to the DataFrame
pc_df["Ram Size"] = pc_df["Ram Size"].apply(map_ram)
```

```python
# Check work
ram_unique_values = pc_df['Ram Size'].unique()
print(f"There are {len(ram_unique_values)} unique Ram Size values: {ram_unique_values}")
```

### SSD Column

```python
# get list of unique values for Ram column
ssd_unique_values = pc_df['SSD Capacity'].unique()
print(f"There are {len(ssd_unique_values)} unique SSD values: {ssd_unique_values}")
```

```python
# Standardize SSD Capacity with help from ChatGPT
# Citation: ChatGPT. (2025, September 9). Conversation with ChatGPT about cleaning and standardizing SSD capacity values in Python/pandas. OpenAI. Personal communication.

def map_ssd(ssd):
    if pd.isna(ssd):
        return "Unknown"
    s = str(ssd).lower().strip()

    # Address missing / unknown values
    if any(x in s for x in ["not included", "not include", "not applicable", 
                            "no ssd", "none", "na", "see options", ".", "0gb", "0"]):
        return "Unknown"

    # Extract all groups of digits from entries, 
    # and replace those without numbers as unknown amount of storage
    numbers = re.findall(r"\d+\.?\d*", s)
    if not numbers:
        return "Unknown"

    sizes = []
    for n in numbers:
        val = float(n)

        # Find and format all TB values
        if "tb" in s:
            val = int(round(val))
            
            # Check for typos; more than 100 TB is non-standard and unlikely for a laptop
            if val > 100:
                return f"{val} GB"
                
            sizes.append(f"{val} TB")

        # Find and format all GB values
        elif "gb" in s:
            val = int(round(val))

            # Normalize known unique values
            # I chose this to maintain the fullness of the dataset. Logically, 
            #    some SSD values might be slightly lower than standard values because
            #    the Ebay user wanted to highlight the amount of USABLE space when
            #    accounting for pre-installed partition of the storage space
            if val == 115:
                val = 120
            if val == 476:
                val = 512

            sizes.append(f"{val} GB")

        else:
            # If no unit label is given, make a logical connection to the number and standard SSD values
            # Since it is highly unlikely that a laptop has 1000 TB SSD, we can safely say that
            #    these are probably GB values and convert the value to corresponding TB values
            if val >= 1000:
                if abs(val - 1024) <= 50: return "1 TB"
                if abs(val - 2048) <= 100: return "2 TB"
                if abs(val - 4096) <= 200: return "4 TB"
                sizes.append(f"{int(val)} GB")
            else:
                # Catch and standardize known unique values
                if int(val) == 115:
                    val = 120
                if int(val) == 476:
                    val = 512
                sizes.append(f"{int(val)} GB")

    # Catch any other values not GB or TB
    if not sizes:
        return "Unknown"

    # Normalize GB-to-TB conversions
    def normalize(v):
        if v.endswith("GB"):
            num = int(v.replace(" GB", ""))
            if num == 1024: return "1 TB"
            if num == 2048: return "2 TB"
            if num == 4096: return "4 TB"
            return f"{num} GB"
        return v

    # If there is a range of values, keep the maximum value
    max_size = max(sizes, key=lambda x: int(re.findall(r"\d+", x)[0]))
    return normalize(max_size)

pc_df["SSD Capacity"] = pc_df["SSD Capacity"].apply(map_ssd)
```

### Maximum Resolution Column

```python
# get list of unique values for Max Res column
max_res_unique_values = pc_df['Maximum Resolution'].unique()
print(f"There are {len(max_res_unique_values)} unique Max Resolution values: {max_res_unique_values}")
```

```python
# Consult ChatGPT to help map low vs high resolution categories and understand what constitudes high vs low res
# Citation: ChatGPT. (2025, September 9). Conversation about sorting laptop screen resolution values into high and low categories, and creating a Python function for cleaning resolution data. OpenAI. Personal communication.

def map_resolution(res):
    if pd.isna(res):
        return "Unknown"
    s = str(res).lower().strip()

    # Regular HD is actually considered low resolution in the computer world
    if "hd" in s and not any(x in s for x in ["full", "fhd", "1920"]):
        return "Low"

    # These are considered high resolution keywords
    if any(x in s for x in ["full", "fhd", "2k", "4k", "uhd"]):
        return "High"

    # Extract numeric width × height values for comparison
    nums = re.findall(r"\d{3,4}", s)
    if len(nums) >= 2:
        width =int(nums[0])
        height = int(nums[1])
        # Rule: Full HD starts at 1920×1080
        if width >= 1920 and height >= 1080:
            return "High"
        else:
            return "Low"

    return "Unknown"

pc_df["Maximum Resolution"] = pc_df["Maximum Resolution"].apply(map_resolution)
```

```python
# Standardize and group HDD Capacity with help from ChatGPT
# Citation: Citation: ChatGPT. (2025, September 9). Conversation about sorting laptop HDD capacity, and creating a Python function for cleaning HDD data. OpenAI. Personal communication.

def map_storage(value):
    if pd.isna(value):
        return "Unknown"
    s = str(value).lower().strip()

    # No Storage
    if any(x in s for x in ["no hdd", "no", "not included", "not applicable", "n.a.", 
        "n/a", "none", "non", "0", "don't apply", "ssd only", "see ssd", "."]):
        return "No Storage"

    # Find all numbers in each value, and label those without numbers as unknown
    nums = re.findall(r"\d+\.?\d*", s)
    if not nums:
        return "Unknown"

    # Catch the values without digits
    try:
        size = float(nums[0])
    except:
        return "Unknown"

    # Convert TB to GB for grouping
    if "tb" in s:
        size_gb = size * 1000
    else:
        size_gb = size

    # Categorize
    if size_gb < 128:
        return "Small (<128 GB)"
    elif 128 <= size_gb <= 512:
        return "Standard (128–512 GB)"
    elif size_gb > 512:
        return "Large (≥1 TB)"
    else:
        return "Unknown"

pc_df["Hard Drive Capacity"] = pc_df["Hard Drive Capacity"].apply(map_storage)
```

```python
# Check work
hdd_unique_values = pc_df['Hard Drive Capacity'].unique()
print(f"There are {len(hdd_unique_values)} unique HDD values: {hdd_unique_values}")
```

### Storage Type Column

```python
# get list of unique values for Hard Drive Capacity column
storage_type_unique_values = pc_df['Storage Type'].unique()
print(f"There are {len(storage_type_unique_values)} unique Storage Type values: {storage_type_unique_values}")
```

```python
# Standardize and group Storage Type with help from ChatGPT
# Citation: ChatGPT. (2025, September 10). Conversation about cleaning and categorizing laptop storage type values (SSD, HDD, Hybrid, eMMC, UFS) using Python functions. OpenAI. Personal communication.

def map_storage_type(x):
    if pd.isna(x):
        return "Unknown"
    s = str(x).strip().lower()

    # Both SSD and HDD
    if ("hdd" in s and "ssd" in s) or "sshd" in s:
        return "Hybrid (SSD+HDD/SSHD)"

    # Only HDD
    if any(k in s for k in ["hdd", "hard drive"]) and "ssd" not in s:
        return "HDD"

    # Flash (embedded) – eMMC / UFS
    if any(k in s for k in ["emmc", "ufs", "eufs", "flash memory"]):
        return "Flash (eMMC/UFS)"

    # SSD only
    if any(k in s for k in ["ssd", "nvme", "m.2", "msata", "solid state"]):
        return "SSD"

    return "Unknown"

pc_df["Storage Type"] = pc_df["Storage Type"].apply(map_storage_type)
```

```python
# Check work
storage_type_unique_values = pc_df['Storage Type'].unique()
print(f"There are {len(storage_type_unique_values)} unique Storage Type values: {storage_type_unique_values}")
```

```python
# Investigate ranges in the Price column
price_series = pd.Series(pc_df['Price'])
ranges=[]
for price in price_series:
    price.strip()
    if 'to' in price or '-' in price:
        ranges.append(price)

ranges.sort()
print(len(ranges))
print(ranges)
```

```python
def map_price(x):
    s = str(x)

    # If a range is present, keep the maximum dollar amount and delete the minimum
    #   Regardless, strip the $ and convert to float so this column can be used as
    #   a continuous variable in analysis
    # Here, I assume that laptop prices do not exceed $999,999
    if "to" in s:
        price_range = s.split("$")
        max_price = price_range[2]
        if ',' in max_price:
            split_max_price = max_price.split(',')
            max_price = split_max_price[0]+split_max_price[1]
        max_price = float(max_price)
        return max_price
    else:
        price = s.lstrip('$')
        if ',' in price:
            split_price = price.split(',')
            price = split_price[0]+split_price[1]
        price = float(price)
        return price

pc_df["Price"] = pc_df["Price"].apply(map_price)
```
